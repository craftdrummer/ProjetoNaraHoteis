# Auditoria do DataFrame Canais de Venda

Este conjunto contém informações sobre os canais utilizados para comercialização das reservas. Nesta etapa será realizada uma análise da estrutura dos dados, identificação de inconsistências e aplicação dos tratamentos necessários para garantir a padronização dos registros.

```python
## Analisando estrutura, tipos de dados e padrões de preenchimento

df_canais_venda.info()
display(df_canais_venda.dtypes)
display(df_canais_venda.describe(include='str'))
df_canais_venda
```

 Problemas identificados no DataFrame Canais de Vendas

| Coluna | Problema | Quantidade | Severidade |
|---------|-----------|------------|------------|
| nome_canal | Padrão de texto | 2 | 🟢 Baixa |
| comissao_pct | Valores armazenados como texto; necessário remover '%'; e converter para float. | 4 | 🟡 Média |

```python
## Tratando inconsistências identificadas durante a auditoria

## Convertendo a comissão para formato numérico adequado para análise
df_canais_venda['comissao_pct'] = (
    df_canais_venda['comissao_pct']
    .astype(str).str.replace('%', '', regex=False)
    .astype(float))
df_canais_venda['comissao_pct'] = (pd.to_numeric(df_canais_venda['comissao_pct'], errors='coerce'))

## Padronizando nomenclaturas dos canais de venda
df_canais_venda['nome_canal'] = (
    df_canais_venda['nome_canal']
    .str.title()
    .replace('Ota', 'OTA'))


## Validando as alterações aplicadas
df_canais_venda.info()
df_canais_venda
```

# Auditoria do DataFrame Tipos de Quarto

Esta base reúne as categorias de acomodação disponibilizadas pela rede. O objetivo da auditoria é verificar a consistência dos valores monetários, identificar duplicidades e padronizar as descrições dos quartos.

```python
## Avaliando estrutura, tipos de dados e qualidade dos registros

df_tipos_quarto.info()
display(df_tipos_quarto.dtypes)
display(df_tipos_quarto.describe(include='str'))
df_tipos_quarto
```

Diagnóstico de Qualidade dos Dados

| Coluna | Problema Identificado | Quantidade de Ocorrências | Severidade |
|---------|----------------------|--------------------------|------------|
| descricao | Falta de padronização na capitalização dos valores textuais | 6 | 🟢 Baixa |
| valor_diaria_base | Armazenamento incorreto de valores monetários como texto, exigindo remoção do prefixo 'R$' e conversão para float. | 4 | 🟡 Média |
| *Registro* | Presença de registros duplicados | 2 | 🔴 Alta |

```python
## Aplicando correções identificadas na etapa de diagnóstico

## Convertendo valores monetários para formato numérico
df_tipos_quarto['valor_diaria_base'] = (
    df_tipos_quarto['valor_diaria_base']
    .astype(str)
    .str.replace('R$', '', regex=False)
    .str.strip()
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False).astype(float)
)
df_tipos_quarto['valor_diaria_base'] = pd.to_numeric(
    df_tipos_quarto['valor_diaria_base'],errors='coerce')

## Removendo duplicatas
df_tipos_quarto = df_tipos_quarto.drop_duplicates()

## Padronizando descrições dos tipos de quarto
df_tipos_quarto['descricao'] = df_tipos_quarto['descricao'].str.title().str.strip()
df_tipos_quarto
```

```python
## Aplicando correções identificadas no DataFrame Unidades

## Padronizando nomenclaturas das regiões
df_unidades['regiao'] = (
    df_unidades['regiao']
    .str.strip()
    .str.title()
    .replace({'Cap': 'Capital'})
)

## Padronizando categorias dos hotéis
df_unidades['categoria_hotel'] = (
    df_unidades['categoria_hotel']
    .str.strip()
    .str.title()
)

mapa = {
    '3*': '3 Estrelas',
    'Três Estrelas': '3 Estrelas',
    'Cinco Estrelas': '5 Estrelas'
}
df_unidades['categoria_hotel'] = df_unidades['categoria_hotel'].replace(mapa)

df_unidades['num_quartos_total'] = (
    df_unidades['num_quartos_total']
    .astype(float)
    .astype('Int64'))

## Corrigindo os valores nulos com os informados pela empresa
correcoes = {
    7: 70,
    10: 65}

for id_unidade, num_quartos_total in correcoes.items():
    df_unidades.loc[
        df_unidades['id_unidade'] == id_unidade,
        'num_quartos_total'
    ] = num_quartos_total

df_unidades = df_unidades.astype({
    'cidade': 'string',
    'regiao': 'string',
    'categoria_hotel': 'string'})


display(df_unidades)
```

# Auditoria do DataFrame Funcionários

Esta etapa busca validar informações cadastrais e funcionais dos colaboradores, incluindo salários, datas de admissão, cargos e departamentos.

## Diagnóstico de Qualidade dos Dados

| Coluna | Problema Identificado | Quantidade de Ocorrências | Severidade |
|---------|----------------------|--------------------------|------------|
| salario | Valores monetários armazenados como texto, exigindo limpeza e conversão para formato numérico | XX | 🟡 Média |
| salario | Presença de valores incompatíveis com o padrão observado na distribuição salarial | XX | 🔴 Alta |
| data_admissao | Tipagem inadequada para análise temporal | XX | 🟡 Média |
| cargo | Falta de padronização textual | XX | 🟢 Baixa |
| departamento | Presença de abreviações e variações de nomenclatura | XX | 🟡 Média |

```python
## Avaliando estrutura e qualidade dos registros cadastrais

df_funcionarios.info()
df_funcionarios
```

```python
## Tratando inconsistências identificadas no cadastro de funcionários

## Convertendo salários para formato numérico
df_funcionarios['salario'] = (
    df_funcionarios['salario']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('.', '', regex=False)
    .str.replace(',', '.', regex=False)
)
df_funcionarios['salario'] = pd.to_numeric(
    df_funcionarios['salario'],
    errors='coerce'
)

display(df_funcionarios)
```

```python
## Convertendo datas de admissão para formato datetime
df_funcionarios['data_admissao'] = pd.to_datetime(
    df_funcionarios['data_admissao'],
    format='mixed',
    dayfirst=True,
    errors='coerce')

print('Tipo:', df_funcionarios['data_admissao'].dtype)
print('NaT (não convertidos):', df_funcionarios['data_admissao'].isnull().sum())
```

```python
## Investigando possíveis valores atípicos na distribuição salarial

print(df_funcionarios[df_funcionarios['salario'] < 0])

Q1 = df_funcionarios['salario'].quantile(0.25)
Q3 = df_funcionarios['salario'].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

outliers = df_funcionarios[
    (df_funcionarios['salario'] < limite_inferior) |
    (df_funcionarios['salario'] > limite_superior)
]

display(outliers)
```

### Análise de Valores Atípicos

Foi realizada uma análise estatística da coluna de salários utilizando o método do Intervalo Interquartil (IQR).

Os registros identificados como potenciais outliers foram avaliados individualmente para verificar se representavam salários válidos ou inconsistências de cadastro.

```python
## Corrigindo registros salariais inconsistentes com os valores informados pela Empresa

correcoes = {
    11: 1950.00,
    26: 2300.00
}

for id_funcionario, salario in correcoes.items():
    df_funcionarios.loc[
        df_funcionarios['id_funcionario'] == id_funcionario,
        'salario'
    ] = salario

display(
    df_funcionarios.loc[
        df_funcionarios['id_funcionario'].isin([11, 26]),
        ['id_funcionario', 'nome', 'salario']
    ]
)
```

```python
## Identificando variações de nomenclatura em cargos e departamentos

display(df_funcionarios['cargo'].value_counts())
display(df_funcionarios['departamento'].value_counts())
```

```python
## Padronizando cargos e departamentos para garantir consistência analítica
df_funcionarios['cargo'] = (
    df_funcionarios['cargo']
    .str.strip()
    .str.title()
)

df_funcionarios['departamento'] = (
    df_funcionarios['departamento']
    .str.strip()
    .str.title()
)

mapa = {
    'A&B': 'Alimentos e Bebidas',
    'Alimentos E Bebidas' : 'Alimentos e Bebidas',
    'Admin.': 'Administração',
    'Gov.': 'Governança'
}
df_funcionarios['departamento'] = df_funcionarios['departamento'].replace(mapa)

display(df_funcionarios['cargo'].value_counts())
display(df_funcionarios['departamento'].value_counts())
```

# Auditoria do DataFrame Reservas

Por se tratar do principal conjunto operacional da empresa, esta auditoria avaliará informações relacionadas às reservas, períodos de hospedagem, avaliações dos hóspedes e métodos de pagamento.

## Diagnóstico de Qualidade dos Dados

| Coluna | Problema Identificado | Quantidade de Ocorrências | Severidade |
|---------|----------------------|--------------------------|------------|
| id_canal | Presença de registros incompatíveis com o tipo de dado esperado | XX | 🟡 Média |
| data_checkin | Tipagem inadequada para operações temporais | XX | 🟡 Média |
| data_checkout | Tipagem inadequada para operações temporais | XX | 🟡 Média |
| qtd_diarias | Quantidade de diárias incompatível com o período registrado | XX | 🔴 Alta |
| avaliacao_hospede | Registros incompatíveis com a escala de avaliação adotada | XX | 🟡 Média |
| num_hospedes | Quantidade de hóspedes inconsistente em algumas reservas | XX | 🟡 Média |
| status_reserva | Falta de padronização textual | XX | 🟢 Baixa |
| forma_pagamento | Presença de abreviações e nomenclaturas divergentes | XX | 🟢 Baixa |

```python
## Avaliando estrutura e qualidade dos registros de reservas

df_reservas.info()
df_reservas
```

```python
## Avaliando indicadores que possam revelar inconsistências operacionais

display(df_reservas[['qtd_diarias', 'num_hospedes']].describe())
display(df_reservas['avaliacao_hospede'].unique()[:20])
```

```python
## Convertendo datas de check-in e check-out para datetime
df_reservas['data_checkin'] = pd.to_datetime(
    df_reservas['data_checkin'],
    errors='coerce'
)

df_reservas['data_checkout'] = pd.to_datetime(
    df_reservas['data_checkout'],
    errors='coerce'
)

## Convertendo id_canal para int
df_reservas['id_canal'] = (
    pd.to_numeric(
        df_reservas['id_canal'],
        errors='coerce'
    )
    .astype('Int64')
)
print('Tipo:', df_reservas['data_checkin'].dtype)
print('Tipo:', df_reservas['data_checkout'].dtype)
```

### Validação da Duração das Reservas

Foi realizada a comparação entre as datas de check-in e check-out e a quantidade de diárias registrada em cada reserva.

Os registros com divergências foram corrigidos para refletir corretamente o período efetivo de hospedagem.

```python
## Corrigindo reservas com quantidade de diárias inconsistente

correcoes = {
    728: 4, 955: 4, 1134: 2,
    1152: 5, 1448: 2, 1581: 2,
    2312: 2, 2384: 2}

for id_reserva, qtd_diarias in correcoes.items():
    df_reservas.loc[
        df_reservas['id_reserva'] == id_reserva,
        'qtd_diarias'
    ] = qtd_diarias

display(df_reservas.loc[df_reservas['id_reserva'].isin([728, 955, 1134, 1152, 1448, 1581, 2312, 2384]),
        ['id_reserva', 'qtd_diarias']
    ]
)
```

```python
## Convertendo avaliações para formato numérico

df_reservas['avaliacao_hospede'] = (
    pd.to_numeric(
        df_reservas['avaliacao_hospede'],
        errors='coerce'
    )
    .astype('Int64')
)

print(df_reservas['avaliacao_hospede'].dtype)

## Corrigindo avaliações incompatíveis com a escala adotada
correcoes = {
    37: 8, 194: 2, 902: 10,
    933: 7, 1538: 3, 1718: 4,
    1866: 3, 2073: 9, 2167: 5, 2210: 2}

for id_reserva, avaliacao_hospede in correcoes.items():
    df_reservas.loc[
        df_reservas['id_reserva'] == id_reserva,
        'avaliacao_hospede'
    ] = avaliacao_hospede

display(df_reservas.loc[df_reservas['id_reserva'].isin([37, 194, 902, 933, 1538, 1718, 1866, 2073, 2167, 2210]),
        ['id_reserva', 'avaliacao_hospede']
    ]
)
```

```python
## Corrigindo registros com quantidade de hóspedes inconsistente

correcoes = {
    568: 2, 675: 3, 903: 2,
    1201: 3, 1883: 1, 2427: 1,}

for id_reserva, num_hospedes in correcoes.items():
    df_reservas.loc[
        df_reservas['id_reserva'] == id_reserva,
        'num_hospedes'
    ] = num_hospedes

display(df_reservas.loc[df_reservas['id_reserva'].isin([568, 675, 903, 1201, 1883, 2427]),
        ['id_reserva', 'num_hospedes']
    ]
)
```

```python
## Avaliando padrões de preenchimento em status e formas de pagamento

display(df_reservas['status_reserva'].value_counts())
display(df_reservas['forma_pagamento'].value_counts())
```

```python

## Padronizando status das reservas e formas de pagamento
df_reservas['status_reserva'] = (
    df_reservas['status_reserva']
    .str.strip()
    .str.title()
)

df_reservas['forma_pagamento'] = (
    df_reservas['forma_pagamento']
    .str.strip()
    .str.title()
)

mapa = {
    'Cancel.': 'Cancelada',
    'Concluida' : 'Concluída',
    'Conf.': 'Confirmada',
    'Cash': 'Dinheiro',
    'C. Crédito': 'Cartão De Crédito', 'Cartao Credito': 'Cartão De Crédito',
    'Cc': 'Cartão De Crédito', 'Cred': 'Cartão De Crédito',
    'Cartao Debito': 'Cartão De Débito', 'Cd': 'Cartão De Débito', 'Déb': 'Cartão De Débito',
    'Ted': 'Transferência', 'Transferencia': 'Transferência'}
df_reservas['status_reserva'] = df_reservas['status_reserva'].replace(mapa)
df_reservas['forma_pagamento'] = df_reservas['forma_pagamento'].replace(mapa)
```
